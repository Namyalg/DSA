```
a = [-2,-3,4,-1,-2,1,5,-3]
```

```
#The kadane algorithm can be used for questions related to finding the max sum of a subarray 

#Any ques of this sort, where u need to find max of a subarray

#The solution takes linear time, although the worst case could be O(n**3) or O(n**2) where u need to iterate over the entire array

#This is a very simple algorithm


#The logic is as explained:


#U have 2 variables one max_end i.e will compute the sum of the elements if it is greater than the zero
#i.e if elements are negative then max_end is set to 0,
#so if the sum/element becomes negative it is set to zero, else it is incremented

#So the values of max_end can be appended to a list and the max can be determined


#Or store it in a list and then print the max value

```

```
max_end = 0
#max_so_far = INT_MIN(a very small value in python)
max_so_far = -293040404
values = []

for i in range(len(a)):
  if max_end < 0:
    max_end = 0
  max_end += a[i]
  values.append(max_end)
  if max_end > max_so_far:
    max_so_far = max_end

print(max_so_far)

print(max(values))
```
